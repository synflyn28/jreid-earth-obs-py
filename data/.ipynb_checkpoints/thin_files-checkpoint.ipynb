```python
import xarray as xr
```

```python
# Thin netCDF
```

```python
# Open file, extract all variables
fname = 'aod/JRR-AOD_v2r0_j01_s201905301853295_e201905301854540_c201905301923100'
aod = xr.open_dataset(fname+".nc")
tmp = list(aod.variables)
```

```python
# List vars to leave in file
keeps = ['AOD550', 'Latitude', 'Longitude']
```

```python
# Remove the "keep" variables from the full var list
encoding_dict = {}
for keep in keeps:
    tmp.remove(keep)
    dict_entry = { keep : {"dtype": "float32","zlib": True, '_FillValue': -999.9} }
    encoding_dict.update(dict_entry)
```

```python
# remove everything else
aod_thinned = aod.drop(tmp)
```

```python
aod_thinned.to_netcdf(fname+"_thinned.nc", engine="netcdf4", encoding=encoding_dict)
```

```python
aod.close()
```

```python
fname = 'thinned_files/3B-HHR.MS.MRG.3IMERG.20160811-S233000-E235959.1410.V06B'
grp_name="Grid/"

imerg = xr.open_dataset(fname+".HDF5", group = grp_name)
tmp = list(imerg.variables)
```

```python
keeps = ['precipitationCal', 'lon', 'lat', 'time', 'precipitationQualityIndex' ]

encoding_dict = {}
for keep in keeps:
    tmp.remove(keep)
    dict_entry = { keep : {"dtype": "float32", "zlib": True, '_FillValue': -9999.9} }
    encoding_dict.update(dict_entry)
```

```python
imerg_thinned = imerg.drop(tmp)
```

```python
coverage = [-94.3, 28.9, -88.8, 33.1]
imerg_thinned_small = imerg_thinned.where(
    (imerg_thinned.lat > 28) & (imerg_thinned.lat < 33) & (imerg_thinned.lon > -94) & (imerg_thinned.lon < -88), drop=True)
```

```python
imerg_thinned_small
```

```python
imerg_thinned_small.to_netcdf(fname+"_thinned.nc", engine="netcdf4", encoding=encoding_dict)
```
